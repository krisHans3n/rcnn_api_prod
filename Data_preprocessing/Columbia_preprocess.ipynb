# Columbia Dataset Preprocessing

<font size=3>Generate the coordinates of the bounding box from the Groundtruth mask.<br>
    We obtained the columbia dataset [here](http://www.ee.columbia.edu/ln/dvmm/downloads/AuthSplicedDataSet/AuthSplicedDataSet.htm)<br>
    In addition, we do not divide dataset into training and test sets. All images in dataset are used for testing.<br>
    We provide text files for the test set. See Readme.md for more details.<font>

```python
import cv2
import os
from glob import glob
import os
```

```python
def bounding_box(image,mask,file):
    num = 0
    box_list=[]
    gray = cv2.cvtColor(mask, cv2.COLOR_BGR2GRAY)
    ret,binary = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)
    contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
    contours = sorted(contours, key=lambda i: len(i),reverse=True)
    x, y, w, h = cv2.boundingRect(contours[0])
    x1=x
    y1=y
    x2=x+w
    y2=y+h
    box_list.append(str(x1)+'_'+str(y1)+'_'+ str(x2)+'_'+str(y2)+'_'+'tamper')
    return box_list,binary
```

```python
def load_image(image_path):
    img = cv2.imread(image_path)
    return img
```

```python
data_path='../dataset/Columbia/'
img_path=data_path+'4cam_splc/'
probe_path=data_path+'/probe/'
mask_path=data_path+'/mask/'

if not os.path.exists(probe_path):
  os.makedirs(probe_path)

if not os.path.exists(mask_path):
  os.makedirs(mask_path)

ext = '*.tif'
filenames = glob(os.path.join(img_path, ext))
print('The number of images is ',len(filenames))

for filename in filenames:
    filename1=[str(i) for i in filename]
    file=''.join(filename1)
    print(file)
    print(img_path+'edgemask/'+file.split('/')[4][:-4]+'_'+'edgemask.jpg')
    mask = load_image(img_path+'edgemask/'+file.split('/')[4][:-4]+'_'+'edgemask.jpg')
    image = load_image(file)
    box_list,mask = bounding_box(image,mask,file)
    for i in range(0,len(box_list)):
         name_str='Tp_'+box_list[i]
    cv2.imwrite(probe_path+name_str.rstrip('_')+ ".png", image)
    cv2.imwrite(mask_path + name_str.rstrip('_') + ".png", mask)
print('\n\n=======================Done==============================!')
```

```python
data_dir = probe_path  # FIXME
ext = 'Tp*'
dataDir = mask_path  # FIXME
cls = ['tamper']

test_name=[]
mani_type=[]

test_filenames = glob(os.path.join(data_dir, ext))
for test_file in test_filenames:
    test_content = os.path.splitext(os.path.basename(test_file))[0].split("_")
    test_name.append(os.path.splitext(os.path.basename(test_file))[0])
    mani_type .append(test_content[-1])


with open(data_path+'test_all_single.txt', 'w') as f:
    for pic in test_name:
        test_content = pic.split("_")
        if test_content[-1] in cls:
            content2 = [str(i) for i in test_content[1:-1]]
            content3=' '.join(content2)
            f.write('%s %s %s\n' % (pic,content3,'tamper'))
        
print("OVER!")

```
